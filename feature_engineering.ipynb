```python
import pandas as pd
```

```python
data = pd.read_csv('wine_quality_class.csv')
data.head()
```

```python
data.corr()['quality rate'].sort_values()
```

```python
X = data.iloc[:,:-2]
y = data['quality rate']
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=20)
model.fit(X,y)
pd.DataFrame(model.feature_importances_,index=X.columns).sort_values(0,ascending=False)
```

```python
data = pd.read_csv('breast_cancer.csv')
X = data.iloc[:,:-1]
y = data.outcome
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=20)
model.fit(X,y)
pd.DataFrame(model.feature_importances_,index=X.columns).sort_values(0,ascending=False)
```
